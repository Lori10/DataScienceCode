## Distribution Plots

To plot the distribution of continuous features we use :
- distplot
- joinplot
- pairplot

To plot the distribution of categorical features we use : 
- boxplot
- violinplot
- countplot
- barplot

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df.head()
```

## Distribution of Continuos Features

## JoinPlot

A join plot allows to study the relationship between 2 numeric variables. The chart display a hexadeciam scatterplot and their histrograms. It is usually a scatterplot, a hexbin plot, a 2D histogram and a 2D density plot. (This is called Bivariate or Multivariate Analysis since we plot and analyze 2 or more features).

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## Pair plot

A pairs plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables. It also plots the histograms of all variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

## Dist plot

Dist plot helps us to check the distribution of the columns feature. It plots the histogram and the kernel density estimator.(This is univariate analysis since we plot and analyze only 1 feature)

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

## Distribution of Categorical Features


## Countplot
Countplot shows the frequency of each category in a categorical feature.

```python
## Count plot

sns.countplot(y='sex',data=df)
```

```python
sns.countplot(x='sex',data=df, hue='smoker')
```

## Barplot
Barplot shows the frequency of each category in a categorical feature.

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
## Bar plot
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
df.head()
```

## Box plot

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot(y='total_bill', data=df)
```

```python
sns.boxplot('smoker','total_bill', data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# categorize my data based on some other categories

sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

## Violin Plot

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(y='total_bill', data=df)
```

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```

```python
sns.violinplot(x="total_bill", y="day", hue='sex', data=df,palette='rainbow')
```
